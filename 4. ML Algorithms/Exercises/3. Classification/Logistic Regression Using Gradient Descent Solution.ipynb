```python
# Loading Libraries

%matplotlib inline

import numpy as np
import random
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
class LogisticRegression:
    
    
    #######################################################################################
    #
    # inputs      : x, theta
    # output      : predicted value of y
    # description : Calculate the dot product of x and theta
    #
    #######################################################################################
    def calLinearEq(self, x, theta):
        
        y = np.dot(x, theta)
        return y
    
    
    #######################################################################################
    #
    # inputs      : x
    # output      : sigmoid of X 
    # description : Calculate the sigmoid function for the value x
    #               sigmoid funtion formula = 1/(1+e^-x)
    #
    #######################################################################################
    def calSigmoid(self, x):
        
        sig = 1/(1 + np.exp(-x))
        return sig
    
    
    #######################################################################################
    #
    # inputs      : x, y, theta0, theta1
    # output      : Square mean error 
    # description : Calculate the square mean error using inputs. x is in vectorized form 
    #               formula = (1 / m) * sum( -y*log(sigmoid(X*theta)) - (1-y)*log( 1 - sigmoid(X*theta)) )
    #
    #######################################################################################
    def calError(self, x, y, theta0, theta1):
        
        error = 0
        
        ## Start writing your code after this block ##
        numOfPoints = len(y)
        res = 0
        for i in range(numOfPoints):
            theta = [theta0, theta1]
            linRes = self.calLinearEq(x[i], theta)
            sig = self.calSigmoid(linRes)
            
            res += (-y[i] * np.log(sig) - (1-y[i])*np.log(1-sig))
       
        return res/float(numOfPoints)
    
    
    #######################################################################################
    #
    # inputs      : x, y, theta0, theta1, alpha
    # output      : new theta0 and theta1 (After calculating the derivation)
    # description : Calculate the step gradient. x is in vectorized form 
    #               
    #
    #######################################################################################
    
    def calStepGradient(self, x, y, theta0, theta1, alpha):
        
        numOfPoints = len(y) 
        new_theta0 = 0
        new_theta1 = 0
        StepGradientTheta = [0, 0]

        for i in range(numOfPoints):
            linRes = self.calLinearEq(x[i], theta)
            sig = self.calSigmoid(linRes)
            StepGradientTheta += x[i] * (sig - y[i])
            
        StepGradientTheta0, StepGradientTheta1 = StepGradientTheta
        new_theta0 = theta0 - ( alpha * StepGradientTheta0 ) / float(numOfPoints)
        new_theta1 = theta1 - ( alpha * StepGradientTheta1 ) / float(numOfPoints)
    
        return new_theta0, new_theta1
    
    def train(self, x, y, theta, alpha, numOfIterations):
        
        # training the model using gradient descent
        theta0 = theta[0]
        theta1 = theta[1]
        #numOfIterations = 1
        for i in range(numOfIterations):
            theta0, theta1 = self.calStepGradient(x, y, theta0, theta1, alpha)
            error = self.calError(x, y, theta0, theta1)
            print("iteration {0} and error {1}".format(i, error))
        return theta0, theta1
    
    def testing(self, x, y, theta0, theta1):

        return self.calError(x, y, theta0, theta1)
```

```python
## hyper parameters
m, n = np.shape(X)
numIterations= 10000
alpha = 0.0001
theta = np.ones(n)
```

```python
#Training the model 
LogisticModel = LogisticRegression()
theta0, theta1 = LogisticModel.train(x_train, y_train, theta, alpha, numIterations)
```

```python
#Testing the model
error = LogisticModel.testing(x_test, y_test, theta0, theta1)
print("testing error is {0}".format(error))
```

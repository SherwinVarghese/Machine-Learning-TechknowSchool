## 2.2 Introduction to relational plots and subplots


### 2.2.1 Introducing relplot()

 Why use `relplot()` instead of `scatterplot()` ?
 - relplot() lets you create subplots in a single â€€figure

<img src="https://drive.google.com/uc?export=view&id=16SHNaIGfVBlOy7SEYnwpoA5GnTvb9GOR" width="90%">


```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter")
plt.show()
```

### 2.2.2 Subplots in columns

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter",
col="smoker")
plt.show()
```

### 2.2.3 Subplots in rows

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter",
row="smoker")
plt.show()
```

### 2.2.4  Subplots in rows and columns

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter",
col="smoker",
row="time")
plt.show()
```

### 2.2.5 Customizing scatter plots

#### 2.2.5.1 Subgroups with point size

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter",
size="size")
plt.show()
```

#### 2.2.5.2 Point size and hue

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter",
size="size",
hue="size")
plt.show()
```

#### 2.2.5.3 Subgroups with point style

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter",
hue="smoker",
style="smoker")
plt.show()
```

#### 2.2.5.4 Changing point transparency

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Set alpha to be between 0 and 1
sns.relplot(x="total_bill",
y="tip",
data=tips,
kind="scatter",
alpha=0.4)
plt.show()
```

##### 2.6.1 Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

mpg = pd.read_csv("https://drive.google.com/uc?export=view&id=1XstyLKwnAaDHOuVW5Vwotqx6fIjMvfm5")
sns.relplot(x="model_year", y="mpg",
data=mpg,
kind="scatter")
plt.show()
```

##### 2.6.2 Line Plot

```python
sns.relplot(x="model_year", y="mpg",
data=mpg,
kind="line")
plt.show()
```

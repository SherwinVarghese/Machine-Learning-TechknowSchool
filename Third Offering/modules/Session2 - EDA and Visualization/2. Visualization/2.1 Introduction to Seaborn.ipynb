### 2.1.2 Tips Dataset

```python
import pandas as pd
import seaborn as sns
tips = sns.load_dataset("tips")
tips.head()
```

### 2.1.3 A basic Scatter plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x="total_bill",
y="tip",
data=tips)
plt.show()
```

### 2.1.3 Specifying hue colors

```python
import matplotlib.pyplot as plt
import seaborn as sns
hue_colors = {"Yes": "black",
"No": "red"}
sns.scatterplot(x="total_bill",
y="tip",
data=tips,
hue="smoker",
palette=hue_colors)
plt.show()
```

### 2.1.3 Using hue with count plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x="smoker",
data=tips,
hue="sex")
plt.show()
```

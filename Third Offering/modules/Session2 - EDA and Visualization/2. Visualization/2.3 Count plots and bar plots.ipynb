## 2.3 Count plots and bar plots


### 2.3.1 countplot() vs. catplot()
 

#### 2.3.1.1 countplot()

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
masculinity_data = pd.read_csv("datasets/masculinity.csv")
sns.countplot(x="how_masculine",
data=masculinity_data)
plt.show()
```

#### 2.3.1.2 catplot()



```python
sns.catplot(x="how_masculine",
data=masculinity_data,
kind="count")
plt.show()
```

### 2.3.2 Changing the order


```python
category_order = ["No answer",
"Not at all",
"Not very",
"Somewhat",
"Very"]
sns.catplot(x="how_masculine",
data=masculinity_data,
kind="count",
order=category_order)
plt.show()
```

### 2.3.3 Bar plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
sns.catplot(x="day",
y="total_bill",
data=tips,
kind="bar")
plt.show()
```

#### 2.3.3.1 Turning off confidence intervals

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
sns.catplot(x="day",
y="total_bill",
data=tips,
kind="bar",
ci=None)
plt.show()
```

### 2.3.4 Changing the orientation

```python
sns.catplot(x="total_bill",
y="day",
data=tips,
kind="bar")
plt.show()
```

### 2.3.5 Box Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
g = sns.catplot(x="time",
y="total_bill",
data=tips,
kind="box")
plt.show()
```

### 2.3.6 Change the order of categories

```python
g = sns.catplot(x="time",
y="total_bill",
data=tips,
kind="box",
order=["Dinner",
"Lunch"])
plt.show()
```

### 2.3.7 Omitting the outliers using `sym`

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
g = sns.catplot(x="time",
y="total_bill",
data=tips,
kind="box",
sym="")
plt.show()
```

### 2.3.8 Changing the whiskers using `whis`

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
g = sns.catplot(x="time",
y="total_bill",
data=tips,
kind="box",
whis=[0, 100])
plt.show()
```

### 2.3.9 Point plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
sns.catplot(x="smoker",
y="total_bill",
data=tips,
hue="sex", #'size',
kind="point")
plt.show()
```
